```python
train_data = pd.read_csv("train.csv")
val_data = pd.read_csv("eval.csv")
test_data = pd.read_csv("test.csv")
```

```python
train_data.head()
```

```python
mapping = {"negative": 0, "fairly negative":1, "neutral": 2, "fairly positive":3, "positive":4}
```

```python
train_data["labels"] = train_data["labels"].map(mapping)
val_data["labels"] = val_data["labels"].map(mapping)
test_data["labels"] = test_data["labels"].map(mapping)
```

```python
result, model_outputs, wrong_preds = model.eval_model(eval_df=val_data, acc=sklearn.metrics.accuracy_score)
```
